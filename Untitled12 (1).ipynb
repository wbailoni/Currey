```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# --- Clients Table ---
clients = []
genders = ['Male','Female']
ethnicities = ['White','Black','Hispanic','Asian','Other']

first_names = ['Alice','Bob','Charlie','Diana','Ethan','Fiona','George','Hannah','Ian','Julia',
               'Kevin','Laura','Mike','Nina','Oscar','Paula','Quentin','Rachel','Steve','Tina',
               'Uma','Victor','Wendy','Xander','Yara','Zane','Liam','Emma','Noah','Olivia',
               'Ava','Sophia','Mason','Logan','Evelyn','Elijah','Harper','James','Isabella','Benjamin',
               'Mia','Lucas','Amelia','Henry','Charlotte','Alexander','Abigail','Daniel','Emily','Jack']

for i in range(1, 101):  # 100 clients
    clients.append({
        'client_id': i,
        'name': random.choice(first_names),
        'age': random.randint(18, 75),
        'gender': random.choice(genders),
        'ethnicity': random.choice(ethnicities)
    })

clients_df = pd.DataFrame(clients)

# --- Programs Table ---
programs = [
    {'program_id':1,'program_name':'Fitness','funder_name':'City Health Dept'},
    {'program_id':2,'program_name':'Nutrition','funder_name':'Community Fund'},
    {'program_id':3,'program_name':'Counseling','funder_name':'Private Donor'},
    {'program_id':4,'program_name':'Wellness','funder_name':'State Grant'},
    {'program_id':5,'program_name':'Rehabilitation','funder_name':'Federal Health'}
]
programs_df = pd.DataFrame(programs)

# --- Client Services Table ---
services = []
service_types = ['Workout','Meal Plan','Therapy','Consultation','Checkup']

service_id = 1
for client in clients_df['client_id']:
    num_services = random.randint(1,5)  # Each client has 1-5 services
    for _ in range(num_services):
        program = random.choice(programs_df['program_id'])
        service_date = datetime.now() - timedelta(days=random.randint(0, 365))
        services.append({
            'service_id': service_id,
            'client_id': client,
            'program_id': program,
            'service_date': service_date.date(),
            'service_type': random.choice(service_types),
            'satisfaction_score': random.randint(1,10)
        })
        service_id += 1

services_df = pd.DataFrame(services)

# --- Optional: Save to CSV ---
clients_df.to_csv('clients.csv', index=False)
programs_df.to_csv('programs.csv', index=False)
services_df.to_csv('client_services.csv', index=False)

print("Mock data created successfully!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Merge data for analysis
df = services_df.merge(clients_df, on='client_id').merge(programs_df, on='program_id')

# 1️⃣ Average Satisfaction by Program
plt.figure(figsize=(8,5))
sns.barplot(x='program_name', y='satisfaction_score', data=df, ci=None)
plt.title('Average Satisfaction by Program')
plt.ylabel('Satisfaction Score')
plt.xlabel('Program')
plt.show()

# 2️⃣ Service Type Count
plt.figure(figsize=(8,5))
sns.countplot(x='service_type', data=df)
plt.title('Number of Services by Type')
plt.ylabel('Count')
plt.xlabel('Service Type')
plt.show()

# 3️⃣ Age Distribution of Clients
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=10, kde=True)
plt.title('Client Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# 4️⃣ Satisfaction by Age Group
df['age_group'] = pd.cut(df['age'], bins=[17,25,35,45,55,65,75], labels=['18-25','26-35','36-45','46-55','56-65','66-75'])
plt.figure(figsize=(8,5))
sns.boxplot(x='age_group', y='satisfaction_score', data=df)
plt.title('Satisfaction Score by Age Group')
plt.ylabel('Satisfaction')
plt.xlabel('Age Group')
plt.show()
```
